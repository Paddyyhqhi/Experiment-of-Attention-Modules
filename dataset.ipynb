```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_pickle('dataset.pkl')

patterns = ['Center','Donut','Edge-Loc','Edge-Ring','Loc','Random','Scratch','Near-full']
all_ = 0
print("Pattern", " "*5, "Numbers of wafers")
print("---"*10)
for i in range(len(patterns)):
    cur = sum(df.failureNum == i)
    print(patterns[i], " "*(20-len(patterns[i])), cur)
    all_ += cur
print("---"*10)
print("Total", " "*14, all_)

fig, ax = plt.subplots()
uni_pattern=np.unique(df.failureType, return_counts=True)
patterns = ['Center','Donut','Edge-Loc','Edge-Ring','Loc','Random','Scratch','Near-full']
ax.bar(patterns,uni_pattern[1]/df.shape[0], color='gold', align='center', alpha=0.9)
ax.set_title("failure type frequency")
ax.set_ylabel("% of pattern wafers")


plt.show()
```
